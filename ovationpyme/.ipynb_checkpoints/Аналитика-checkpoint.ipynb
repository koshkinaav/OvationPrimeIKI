```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import ttest_ind
```

**Посколько процесс подсчета модели не простой для компьютера, а данные представлены за десятилетие, я пока подготовил только предсказания за 2016 год за каждый ЧАС**

В первой таблице собраны предсказания интегрального для 4 типов частицов: diffuse, monoenergetic, wave, ions.


Данные получены подменой "оригинальных" OMNI данных  https://spdf.gsfc.nasa.gov/pub/data/omni/ историческими для 
спутников ACE, WIND, DSCOVR, которые находятся по следующему адресу: 


https://spdf.gsfc.nasa.gov/pub/data/omni/high_res_omni/

```python
df = pd.read_csv('flux_forecasts.csv').rename(columns={'dt': 'Epoch'})
df['Epoch'] = pd.to_datetime(df['Epoch'])
```

```python
df
```

**В следющей таблице находятся, посчитанные значения Newell Coupling Function для каждого спутника. Для интереса я добавил Kp индекс.**

```python
newell = pd.read_csv('Newell_results.csv')[['Epoch','newell_ace', 'newell_wind', 'newell_dscovr', 'newell', 'Kp']]
newell['Epoch'] = pd.to_datetime(newell['Epoch'])
merged_df = df.merge(newell, on=['Epoch'])
```

```python
merged_df.head()
```

**Посмотрим, например, на 2016-12-31**

```python
start_date = '2016-12-31'
end_date = '2017-01-01'
```

```python
df = merged_df[(merged_df['Epoch'] >= start_date) & (merged_df['Epoch'] < end_date)]
```

```python
df.head()
```

```python
df['hour'] = df['Epoch'].apply(lambda x: x.hour)
```

```python
plt.xlabel('Hour')
plt.ylabel('Newell coupling Fuction')
plt.plot(df['hour'], df['newell'], color='black', label='newell_model' )
plt.plot(df['hour'], df['newell_ace'], color='green', label='newell_ace' )
plt.plot(df['hour'], df['newell_dscovr'], color='blue', label='newell_dscovr' )
plt.plot(df['hour'], df['newell_wind'], color='red', label='newell_wind' )
plt.legend()
```

**Интересно провести t-тест на проверку равенства средних для newell и newell_ace**

```python
newell = np.array(merged_df[~(merged_df.newell.isna())& ~(merged_df.newell_ace.isna())].newell)
newell_ace = np.array(merged_df[~(merged_df.newell.isna())& ~(merged_df.newell_ace.isna())].newell_ace)
```

```python
ttest_ind(newell, newell_ace, equal_var=True)
```

Не отклоняем гиоптезу о равенстве средних, что кончено логично

Интересно посмотреть какой средний результат по модели для Kp индекса

```python
merged_df.groupby('Kp')[['newell', 'newell_ace', 'newell_dscovr', 'newell_wind']].mean()
```

Можно заметить, что в среднем различия в значениях newell coupling function могут достигать 15%, что, вообще говоря, вполне могло привести к ложно отрицательному прогнозу полярного сияния  

**Теперь можно посмотреть предсказания авроальног потока diffuse aurora на северном полюсе для каждого из спутников например за 2016-12-31**

```python
df = df[df.type == 'diff']
```
